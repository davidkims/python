```python
import numpy as np
import matplotlib.pyplot as plt

# Effects of different scalars

# a list of scalars:
scalars = [1, 2, 1/3, 0, -2/3]

baseVector = np.array([0.75, 1])

# Set the font to Arial
plt.rcParams['font.family'] = 'Arial'

# create a figure
fig, axs = plt.subplots(1, len(scalars), figsize=(12, 3))
i = 0  # axis counter

for s in scalars:
    # compute the scaled vector
    v = s * baseVector

    # plot it
    axs[i].arrow(0, 0, baseVector[0], baseVector[1], head_width=0.3, width=0.1, color='k', length_includes_head=True)
    axs[i].arrow(0.1, 0, v[0], v[1], head_width=0.3, width=0.1, color=[0.75, 0.75, 0.75], length_includes_head=True)
    axs[i].grid(linestyle='--')
    axs[i].axis('square')
    axs[i].axis([-2.5, 2.5, -2.5, 2.5])
    axs[i].set(xticks=np.arange(-2, 3), yticks=np.arange(-2, 3))
    axs[i].set_title(f'Scalar = {s:.2f}')
    axs[i].set_xlabel('X-axis')
    axs[i].set_ylabel('Y-axis')
    i += 1  # update axis counter

plt.tight_layout()
plt.savefig('Figure_02_03.png', dpi=300)
plt.show()
```

```python
import os
import shutil

# Find and remove __pycache__ directories recursively
def clear_pycache(path='.'):
    for root, dirs, files in os.walk(path):
        for dir in dirs:
            if dir == '__pycache__':
                pycache_path = os.path.join(root, dir)
                shutil.rmtree(pycache_path)
                print(f"Removed: {pycache_path}")

# Clear __pycache__ directories in the current directory
clear_pycache()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# The vectors
v = np.array([1, 2])
w = np.array([4, -6])
vPlusW = v + w

# now plot all three vectors
plt.figure(figsize=(6, 6))

a1 = plt.arrow(0, 0, v[0], v[1], head_width=0.3, width=0.1, color='k', length_includes_head=True)
a2 = plt.arrow(v[0], v[1], w[0], w[1], head_width=0.3, width=0.1, color=[0.5, 0.5, 0.5], length_includes_head=True)
a3 = plt.arrow(0, 0, vPlusW[0], vPlusW[1], head_width=0.3, width=0.1, color=[0.8, 0.8, 0.8], length_includes_head=True)

# make the plot look a bit nicer
plt.grid(linestyle='--', linewidth=0.5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([a1, a2, a3], ['v', 'w', 'v+w'])
plt.title('Vectors v, w, and v+w')

plt.savefig('Figure_02_02a.png', dpi=300)  # write out the fig to a file
plt.show()
```

```python
# the function
def normOfVect(v):
  return np.sqrt(np.sum(v**2))

# test it on a unit-norm vector
w = np.array([0,0,1])
print( normOfVect(w) )

# non-unit-norm vector, and confirm using np.linalg.norm
w = np.array([1,2,3])
print( normOfVect(w),np.linalg.norm(w) )
```

```python
import os
import shutil

def get_size(path='.'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)
            total_size += os.path.getsize(filepath)
    return total_size

def check_cache_size(path='.'):
    cache_path = os.path.join(path, '__pycache__')

    if not os.path.exists(cache_path):
        print("Cache directory not found.")
        return

    cache_size = get_size(cache_path)

    if cache_size == 0:
        print("Cache directory is empty.")
    else:
        print(f"Cache directory size: {cache_size / (1024 ** 2):.2f} MB")

# Specify the directory path, or use the default (current directory)
directory_path = '/path/to/your/directory'  # Change this to your desired directory
check_cache_size(directory_path)
```

캐시 데이터 체크용 코드
(데이터분석 많이 진행했을 경우에 한번더 체크 할것)

```python
# define function
def createUnitVector(v):
  # get vector norm
  mu = np.linalg.norm(v)
  # return unit vector
  return v / mu


# test on a unit vector
w = np.array([0,1,0])
print( createUnitVector(w) )

# test on a non-unit vector that is easy to confirm
w = np.array([0,3,0])
print( createUnitVector(w) )

# test on a non-unit vector
w = np.array([13,-5,7])
uw = createUnitVector(w)
print( uw )
```
